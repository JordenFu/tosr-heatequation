### 2.1 Background

The broad goal of the coursework is to use the techniques we have learned in the course to study the solutions to a particular non linear partial differential equation called [the heat equation](https://en.wikipedia.org/wiki/Heat_equation). Financially minded people may also want to note that this is almost exactly the [Black-Scholes equation](https://en.wikipedia.org/wiki/Black–Scholes_model) that governs derivatives, but in this coursework we will think of our problem as a model of temperature along a 1d bar of length $L$ (in degrees celsius).

Once complete, the notebook should provide a self-contained report on your numerical investigations and provide relevant commentary in markdown cells around it that would enable another student to understand, run and make basic modifications to the notebook.

The heat equation we will solve is:

$ \frac{\partial T}{\partial t} = \alpha \frac{\partial T^2}{\partial x^2} - \beta (T^4 - S^4) = 0, \quad (1)$

Here $T(x,t)$ is the time dependent temperature profile along the 1d bar that we want to solve for and $S(x,t)$ is a function that represents the temperature of the surroundings, which provides a source or sink of heat if its value is different to the temperature of our object.

To solve the problem we will need to specify the function $S(x,t)$. This will have the form:

$ S(x,t) = A + B \cos(\omega t) e^{-(x/C)^2}  \quad (2)$ 

We will also need the initial Temperature profile:

$ T(x,t=0) = T_0 (1 + e^{-x^2})  \quad (3) $ 

And we will assume that the boundary conditions are such that 

$ T(x=-L/2) = T(x=L/2) = T_0 $

Unless the question indicates otherwise, you should use the values: 

$A = 20.0$, $B = 1.0$, $C = 1.5$, $\omega = 0.5$, $T_0 = 20$, $\alpha = 0.5$, $\beta=0.0001$ and $L=10$.

```python
class Surrounding_function_S:
    def __init__(self, A = 20.0, B = 1.0, omega = 0.5, C = 1.5):
        

        self.A = A
        self.B = B
        self.C = C
        self.omega = omega
    
    def spatial_profile(self, x, t):
        return self.A + self.B * np.cos(self.omega * t) * np.exp(-((x / self.C) ** 2))
    
    def plot_s_of_t_with_fix_x(self, x, max_time):
        # Defensive Programming
        assert type(x) == np.ndarray, "x range should be a numpy array"
        assert x.shape[0] == 1, "The input shape of the fixed x function should be 1."
        assert max_time > 0, "Maximum time should be positive."
        assert type(max_time) == int, "Input time should be integer"
        
        t_range = np.linspace(0, max_time, 100)
        x_result = np.zeros((t_range.shape[0], 1))
        for i, t_i in enumerate(t_range):
            x_result[i] = self.spatial_profile(x, t_i)
        
        plt.plot(t_range, x_result, label = f"x = {x[0]}")
        plt.xlabel('t')
        plt.ylabel(f'S({x[0]}, t)')
        plt.ylim(18.9, 21.1)
        plt.title(f'spatial profile S(t) at x = {x[0]}')
        plt.legend()
        plt.grid()

    def plot_s_of_t_with_range_x(self, x_range, max_time):
        # Defensive Programming
        assert type(x_range) == np.ndarray, "x range should be a numpy array"
        assert max_time > 0, "Maximum time should be positive."
        assert type(max_time) == int, "Input time should be integer"

        t_range = np.linspace(0, max_time, 100)
        x_result_with_t = np.zeros((t_range.shape[0], x_range.shape[0]))
        plt.figure(figsize = (10, 6))
        for i, t_i in enumerate(t_range):
            x_result_with_t[i, :] = self.spatial_profile(x_range, t_i)
        
        for i, x_i in enumerate(x_range):
            plt.plot(t_range, x_result_with_t[:, i], label = f'x={x_i}')
        plt.xlabel('t')
        plt.ylabel('value over time of t')
        plt.xlim(0, max_time)
        plt.title('spatial profile S(t)')
        plt.legend(bbox_to_anchor=(1.15, 1))
        plt.tight_layout()
        plt.grid()

    def plot_s_of_x_with_fix_t(self, x_range, time):
        # Defensive Programming
        assert type(x_range) == np.ndarray, "x range should be a numpy array"
        assert time >= 0, "Maximum time should be positive."
        assert type(max_time) == int, "Input time should be integer"
        
        x_result_with_t = self.spatial_profile(x_range, time)

        plt.figure(figsize = (10, 6))
        plt.plot(x_range, x_result_with_t, label = f"t = {time}")
        plt.xlabel('x')
        plt.ylabel(f'S(x, {time})')
        plt.title('spatial profile S(x)')
        plt.legend()
        plt.grid()
    

    def plot_s_of_x_with_range_t(self, x_range, max_time):
        # Defensive Programming
        assert type(x_range) == np.ndarray, "x range should be a numpy array"
        assert max_time > 0, "Maximum time should be positive."
        assert type(max_time) == int, "Input time should be integer"

        t_range = np.linspace(0, max_time, max_time + 1)
        x_result_with_t = np.zeros((t_range.shape[0], x_range.shape[0]))
        plt.figure(figsize = (10, 6))
        for i, t_i in enumerate(t_range):
            x_result_with_t[i, :] = self.spatial_profile(x_range, t_i)
        
        for i, t_i in enumerate(t_range):
            plt.plot(x_range, x_result_with_t[i, :], alpha=(1-t_i/max_time), label = f"t = {t_i}")
        plt.xlabel('x')
        plt.ylabel('value over time of t')
        plt.title('spatial profile S(x)')
        plt.legend(bbox_to_anchor=(1.05, 1.0))
        plt.grid()
    


# Below are the tests for the class.

my_s_function = Surrounding_function_S()
# x_fix as the input for the fixed x method
x_fix = np.array([0])
# x range as the input for s_of_x_with_fix_t method
x_for_s_of_t = np.linspace(-5, 5, 11)
# x range as the input for s_of_x_with_fix_t method
x_for_s_of_x = np.linspace(-5, 5, 100)
# maximum time
max_time = 20


my_s_function.plot_s_of_t_with_fix_x(x_fix, max_time)
# my_s_function.plot_s_of_t_with_range_x(x_for_s_of_t, max_time)
# my_s_function.plot_s_of_x_with_fix_t(x_for_s_of_x, 20)
# my_s_function.plot_s_of_x_with_range_t(x_for_s_of_x, max_time)
my_s_function.spatial_profile(3, np.pi)
```

In this section, I first construct the polynomial using `SymPy` functions. Firstly, The process begins by setting up the four points to be used in the construction by utilizing the `spatial_profile` method. Next, defining the symbol $x$ as the variable for the polynomial $P(x)$. With these preparations, I can begin building the equation.  

The basic function of the **Lagrange polynomial** is:
$$L_k(x) = \prod_{\substack{i=0, i \neq k}}^{n} \frac{(x - x_i)}{(x_k - x_i)}$$
Finaly, their weights are the values of the functions at each point, resulting in the **Lagrange interpolant**:
$$P_n(x) = \sum_{k=0}^{n} L_k(x) f(x_k)$$

Using the four points $(-2.0, 20.09131828), (-0.5, 20.48348375),   (0.2, 20.53078181), (1.5, 20.19876611)$, the Lagrange interpolant is given by:
$$P_4(x) = \frac{(x - x_2)(x - x_3)(x - x_4)}{(x_1 - x_2)(x_1 - x_3)(x_1 - x_4)} f(x_1) + \frac{(x - x_1)(x - x_3)(x - x_4)}{(x_2 - x_1)(x_2 - x_3)(x_2 - x_4)} f(x_2) + \frac{(x - x_1)(x - x_2)(x - x_4)}{(x_3 - x_1)(x_3 - x_2)(x_3 - x_4)} f(x_3) + \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_4 - x_1)(x_4 - x_2)(x_4 - x_3)} f(x_4)$$
The resulting polynomial is printed below.

```python
# Construct the polynomial using sympy
x_points = np.array([-2.0, -0.5, 0.2, 1.5])
y_points = my_s_function.spatial_profile(x_points, 2)

# Setup the points for convenience
[x1, x2, x3, x4] = x_points
[s1, s2, s3, s4] = y_points

# Compue the polynomial
x = symbols('x')
L1 = (x - x2) / (x1 - x2) * (x - x3) / (x1 - x3) * (x - x4) / (x1 - x4)
L2 = (x - x1) / (x2 - x1) * (x - x3) / (x2 - x3) * (x - x4) / (x2 - x4)
L3 = (x - x1) / (x3 - x1) * (x - x2) / (x3 - x2) * (x - x4) / (x3 - x4)
L4 = (x - x1) / (x4 - x1) * (x - x2) / (x4 - x2) * (x - x3) / (x4 - x3)
P = s1 * L1 + s2 * L2 + s3 * L3 + s4 * L4
print("Using sympy we get", simplify(P))
```

To verify the result, I use the `scipy.interpolate.lagrange` function to compute the Lagrange polynomial. From the output, we can observe that the result matches the polynomial constructed using the `scipy` method.

Next, I compare the result of the **Lagrange interpolant** method (using four points) with the true function. I plot the Lagrange polynomial, the four points, and the true function on the same figure to visualize and analyze the differences. ( To produce a smoother plot, I set a range `x_true` with 100 points. This provides a higher resolution for the plot. )

```python
# Check the implementation using scipy
lagrange_polynomial = lagrange(x_points, y_points)
print("Using the scipy function we get \n",lagrange_polynomial)

# Plot the solution in the range [-5, 5]
a = -5
b = 5
x_true = np.linspace(a, b, 100)
plt.plot(x_true, my_s_function.spatial_profile(x_true, 2), '-', label="true function")
plt.plot(x_true, lagrange_polynomial(x_true), '--', label = "Lagrange polynomial")
plt.plot(x_points, y_points, 'o', label="points given")
plt.title("Comparison from x = -5 to x = 5")
plt.xlabel("x")
plt.ylabel("S(x)")
plt.legend()
plt.grid();

# Plot the solution in the range [-3, 3] for better observation
plt.figure()
a = -3
b = 3
x_true = np.linspace(a, b, 100)
plt.plot(x_true, my_s_function.spatial_profile(x_true, 2), '-', label="true function")
plt.plot(x_true, lagrange_polynomial(x_true), '--', label = "Lagrange polynomial")
plt.plot(x_points, y_points, 'o', label="points given")
plt.title("Comparison from x = -3 to x = 3")
plt.xlabel("x")
plt.ylabel("S(x)")
plt.legend()
plt.grid();
```

**Conclusion**  
According to these figures, we can observe that using only four points to fit is not accurate. The figure illustrates that the interpolation is resonably precise within the range $ x \in [-2, 1.5] $, which is the range of given points. However, outside this range, there is a significant difference between the Lagrange polynomial and the true function. **Conclusion**:  
As the number of points increases from 6 to 9, the Lagrange polynomial provides a better approximation of the true function within the interior (this can be verified by plotting the "true function" for comparison). However, using equally spaced points lead to spurious oscillations at the edges of the interval, which is known as **Runge’s phenomenon**. This results in worse accuracy at the boundaries using higher resolution.

#### 2.2.3. Q3 Differentiation (12 marks total for 2 parts (a)-(b)):

The code to generate the pseudospectral derivative matrices for the first and second derivatives of any function $f(x)$ for a grid containing $N$ points *at the Gauss Lobato points* in $x$ is given below for the standard interval $[-1,1]$, and verified on some example functions.

```python
# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]
# (For use in part 2.2.3)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]

def get_pseudospectral_first_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == 0 and idx_j == 0) : 
                D_matrix[idx_i,idx_j] = (2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == N and idx_j == N) :
                D_matrix[idx_i,idx_j] = -(2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == idx_j) :
                D_matrix[idx_i,idx_j] = - u_i[idx_i] / 2.0 / (1.0 - u_i[idx_i] * u_i[idx_i])
                
            else :
                D_matrix[idx_i,idx_j] = (c_i[idx_i] / c_i[idx_j] * (-1)**(idx_i+idx_j) 
                                         / (u_i[idx_i] - u_i[idx_j]))
    # Fix numerical errors when function flat         
    for idx_i in indices :
        D_matrix[idx_i,idx_i] = 0
        for idx_j in indices :
            if (idx_j != idx_i) : 
                D_matrix[idx_i,idx_i] += -D_matrix[idx_i,idx_j] 
    return D_matrix

def get_pseudospectral_second_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D2_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if((idx_i == idx_j) and (idx_i == 0 or idx_i == N)) : 
                D2_matrix[idx_i,idx_j] = (N ** 4.0 - 1.0) / 15.0
            
            elif ((idx_i == idx_j) and (idx_i != 0) and (idx_i != N)) :
                D2_matrix[idx_i,idx_j] = (- ((N*N - 1.0) * (1.0 - u_i[idx_i] * u_i[idx_i]) + 3.0) 
                                             / (3.0 * (1.0 - u_i[idx_i] * u_i[idx_i]) ** 2.0))
                 
            elif (idx_i == 0 and idx_j > 0) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** idx_j) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 - u_i[idx_j]) - 6.0)
                                          / (1.0 - u_i[idx_j]) / (1.0 - u_i[idx_j]))
                
            elif (idx_i == N and idx_j < N) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** (N+idx_j)) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 + u_i[idx_j]) - 6.0)
                                          / (1.0 + u_i[idx_j]) / (1.0 + u_i[idx_j]))
                
            else :
                D2_matrix[idx_i,idx_j] = (((-1) ** (idx_i+idx_j)) / c_i[idx_j] 
                                            * (u_i[idx_i]*u_i[idx_i] + u_i[idx_i]*u_i[idx_j] - 2.0)
                                            / (1.0 - u_i[idx_i]*u_i[idx_i]) / (u_i[idx_i] - u_i[idx_j])
                                            / (u_i[idx_i] - u_i[idx_j]))
                

    # Fix numerical errors when function flat            
    for idx_i in indices :
        D2_matrix[idx_i,idx_i] = 0
        for idx_j in indices :
            if (idx_j != idx_i) : 
                D2_matrix[idx_i,idx_i] += -D2_matrix[idx_i,idx_j]
                
    return D2_matrix
```

1. `get_FD_first_derivative_matrix(a_N)`: To construct the fourth order accurate finite difference stencil for the 1st derivative, we require **5** points in each stencil. The coefficients are calculated using the **Lagrange polynomial interpolation method**, which involves the following steps:
    * Build the interpolating polynomial using the given points.  
        $$L_k(x) = \prod_{\substack{i=-2 \\ i \neq k}}^{2} \frac{(x - x_i)}{(x_k - x_i)}$$
    * Differentiate the polynomial to obtain the first derivative.
    * Evaluate the derivative at $x=0$.  

To calculate the coefficient $c_{0}$ as an example, I start with the Lagrange polynomial for $c_{0}$, which is defined as:
$$ L_0(x) = \frac{(x - x_{-2})(x - x_{-1})(x - x_{1})(x - x_{2})}
{(x_{0} - x_{-2})(x_{0} - x_{-1})(x_{0} - x_{1})(x_{0} - x_{2})}$$
Here, the points are defined as:
$$x_{-2} = x - 2h, \quad x_{-1} = x - h, \quad x_{1} = x + h, \quad x_{2} = x + 2h$$

Substituting these values into the equation gives:
$$L_0(x) = \frac{(x - (-2h))(x - (-h))(x - h)(x - 2h)}
{(2h)(h)(-h)(-2h)}$$

Next, we differentiate the polynomial to obtain:
$$L_0'(x) = \frac{4x^3 - 10h^2x}{4h^4}$$

Finally substitute $x=0$, we can get the coefficient $c_{0} = 0$.  
Using the same procedure, we can construct the fourth-order accurate finite difference stencil for the first derivative as:
$$ \frac{1}{12h} [ -f(x - 2h) + 8f(x - h) - 8f(x + h) + f(x + 2h)]$$

2. `get_FD_second_derivative_matrix(a_N)`: To construct the second order accurate finite difference stencil for the 2nd derivative, we require **3** points in each stencil. Following a similar process as described above,  but taking the second derivative of the Lagrange polynomial, we can construct the stencil as:
$$\frac{1}{h^{2}}[f(x - h) - 2f(x) + f(x + h)]$$

**Construct the Stencil**  
Next, I will begin constructing the matrix for the stencil in the code. To construct the first derivative stencil matrix, the element $M_{ij}$ is the coefficient to multiply for $x_{j}$ when computing $f'(x_{i})$. Therefore, $f'(x_{i})$ can be expressed as:
$$f'(x_{i}) = M_{i,j-2} \times f(x_{j-2}) + M_{i,j-1} \times f(x_{j-1}) +M_{i,j} \times f(x_{j}) +M_{i,j+1} \times f(x_{j+1}) +M_{i,j+2} \times f(x_{j+2})$$
Based on the discussion above, the coefficients are derived as:
$$M_{i,j-2} = -\frac{1}{12h}, \quad M_{i,j-1} = \frac{8}{12h}, \quad M_{i,j} = 0, \quad M_{i,j+1} = -\frac{8}{12h}, \quad M_{i,j+2} = \frac{1}{12h}$$

For the edge points ($i = 0, 1, n-1, n$), there are not enough points to compute the derivative, as indices such as $i = -2, -1, n+1 n+2$ do not exist. In this case, I use 3-point and 4-point stencil to calulate (e.g., $i=0, 1, 2$ for $f'(x_{0})$). Referring to the [finite difference calculator](https://web.media.mit.edu/~crtaylor/calculator.html), we can obtain these coefficients. A similar procedure is applied for the second derivative using 3-point stencils. (Notice: A 2-point stencil cannot be used to calculate the second derivative at the edge points. Therefore, a 3-point stencil is used instead.)

```python
def get_FD_first_derivative_matrix(a_N) :
    assert a_N > 0, "Simulation points need to be greater than zero."

    # [-1, 1] is the full range
    h = 2.0 / a_N
    one_over_h = 1.0 / h
    D_matrix = np.zeros([a_N + 1,a_N + 1])

    indices = np.arange(a_N + 1)
    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == idx_j) :
                D_matrix[idx_i,idx_j] = 0.0/12.0 * one_over_h

            elif (idx_i == (idx_j+2)) :
                D_matrix[idx_i,idx_j] = 1.0/12.0 * one_over_h
                
            elif (idx_i == (idx_j+1)) :
                D_matrix[idx_i,idx_j] = -8.0/12.0 * one_over_h
                
            elif (idx_i == (idx_j-1)) :
                D_matrix[idx_i,idx_j] = 8.0/12.0 * one_over_h

            elif (idx_i == (idx_j-2)) :
                D_matrix[idx_i,idx_j] = -1.0/12.0 * one_over_h
                
            else :
                D_matrix[idx_i,idx_j] = 0.0

    # Fix the ends with one sided derivatives
    D_matrix[0,0] = -3.0/2.0 * one_over_h
    D_matrix[0,1] = 4.0/2.0 * one_over_h
    D_matrix[0,2] = -1.0/2.0 * one_over_h
    D_matrix[1,0] = -2.0/6.0 * one_over_h
    D_matrix[1,1] = -3.0/6.0 * one_over_h
    D_matrix[1,2] = 6.0/6.0 * one_over_h
    D_matrix[1,3] = -1.0/6.0 * one_over_h
    D_matrix[a_N - 1,a_N-3] = 1.0/6.0 * one_over_h
    D_matrix[a_N - 1,a_N-2] = -6.0/6.0 * one_over_h
    D_matrix[a_N - 1,a_N-1] = 3.0/6.0 * one_over_h
    D_matrix[a_N - 1,a_N-0] = 2.0/6.0 * one_over_h
    D_matrix[a_N,a_N] = 1.0/2.0 * one_over_h
    D_matrix[a_N,a_N-1] = -4.0/2.0 * one_over_h
    D_matrix[a_N,a_N-2] = 3.0/2.0 * one_over_h
    
    
    return D_matrix

# Your code here...

def get_FD_second_derivative_matrix(a_N) :
    assert a_N > 0, "Simulation points need to be greater than zero."
    
    # [-1, 1] is the full range
    h = 2.0 / a_N
    one_over_h2 = 1.0 / (h * h)
    D2_matrix = np.zeros([a_N + 1,a_N + 1])

    indices = np.arange(a_N + 1)
    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == idx_j) :
                D2_matrix[idx_i,idx_j] = -2.0 * one_over_h2
                
            elif (idx_i == (idx_j+1)) :
                D2_matrix[idx_i,idx_j] = 1.0 * one_over_h2
                
            elif (idx_i == (idx_j-1)) :
                D2_matrix[idx_i,idx_j] = 1.0 * one_over_h2
                
            else :
                D2_matrix[idx_i,idx_j] = 0.0

    # Fix the ends with one sided derivatives
    D2_matrix[0,0] = 1.0 * one_over_h2
    D2_matrix[0,1] = -2.0 * one_over_h2
    D2_matrix[0,2] = 1.0 * one_over_h2
    D2_matrix[a_N,a_N] = 1.0 * one_over_h2
    D2_matrix[a_N,a_N-1] = -2.0 * one_over_h2
    D2_matrix[a_N,a_N-2] = 1.0 * one_over_h2
    
    
    return D2_matrix
```

**Classification of second order PDEs**  
To classify the type of PDE, consider the most general second order PDE for 1 dependent variable with 2
independent variables: 
$$A \frac{\partial^2 u}{\partial x^2} + 2B \frac{\partial^2 u}{\partial x \partial y} + C \frac{\partial^2 u}{\partial y^2} + D \frac{\partial u}{\partial x} + E \frac{\partial u}{\partial y} + F = 0$$
The equation is classified based on the discriminant:
$$\Delta = B^2 - 4AC$$
* The equation is a **hyperbolic equation** if  $\Delta > 0$  
* A **parabolic equation** if $\Delta = 0$  
* An **elliptic equation** if $\Delta < 0$  

The PDE Eq. (1) can be written as:
$$\alpha \frac{\partial T^2}{\partial x^2} - \frac{\partial T}{\partial t} - \beta (T^4 - S^4) = 0$$
Therefore, the coefficients are $A = \alpha$, $B = C = 0$, $\Delta = 0$, indicating that Eq. (1) is a **parabolic equation**. More specifically, it is a **Linear Parabolic equation**, meaning this equation is guaranteed to be well-posedness. (can be proven using theorems in mathematics).  

To solve the PDE equation, we start by obtaining the solution using the initial condition at $t=0$. Then, for each subsequent time $t$, we use the solution from the previous time step to compute the results iteratively. This approach involves using **finite equally spacing stencil** or **Gauss-Lebato nodes** and **time-stepping schemes** to compute the solution.

To solve the equation, I construct a class `HeatEquationPDE` that allows us to organize and utilize the required parameters conveniently. his class sets up various parameters, including `alpha`, `beta` for us to define an build-in attribute `Surrounding_function_S` for source term in the partial differential equation. Default values are provided for these parameters, so if there are no specific requirements, the default values will be used.

Required Inputs for the Class:
1. `x_points` : for the spatial domain where the simulation will take place.
2. `initial_value` : The initial temperature profile at $t = 0$, calculated using the equation:
$$T(x,t=0) = T_0 (1 + e^{-x^2})$$
3. `max_time` : The maximum time fot the simulation.
4. `CFL` : The CFL number ensures stability during the simulation when using the RK2 time integration method. This value is typically smaller $1/\lambda$, but the precise value must be tested for each case. Details will be discussed in the results section.
5. `method` : The method for computing the second order differential of x. There are two methods:
    * `finite_difference` : Using a 3-points equally spacing stencil to compute.
    * `pseudospectral` : Using a 5-point Gauss-Lobatto node stencil to compute.  
    * The stencils using for both methods are defined in Q3.

`calculate_dydt` : 
To construct the PDE, we use the equation:
$$\frac{\partial T}{\partial t} = \alpha \frac{\partial T^2}{\partial x^2} - \beta (T^4 - S^4)$$
where $S$ is directly the class `Surrounding_function_S` class defined in Q2. For stability, the boundary values at $x = -5$ and $x = 5$ are set to be 0 when using the `pseudospectral` method for simulation.  

Finally, there are two methods to solve the PDE:
1. `integrate_using_solve_ivp` : This method uses `scipy.integrate.solve_ivp` to simulate the results up to `max_time`. The time step is fixed with 100 points.  
2. `integrate_using_midpoint` : This method uses the RK2 time integration (midpoint) method for simulation. The time step $\Delta t$ is calculated as $\text{CFL} \times \Delta x$, meaning the time step increases as the CFL number increases.

For `HeatEquationPDE` class, I insert one additional assert defensive programming:
1. When using the `finite_difference` method, the input `x_points` must consist of equally spaced points.  
2. When using the `pseudospectral` method, the input `x_points` must be Gauss-Lobato nodes.

```python
# First make a class for the Heat Equation

class HeatEquationPDE :

    
    # constructor function
    def __init__(self, x_points, initial_value, max_time, CFL = 1.0, method = 'finite_difference', alpha=0.5, beta=0.0001, A=20.0, B=1.0, C=1.5, omega=0.5):
        
        # Defensive Programming
        assert omega >= 0, "Frequency cannot be negative"
        assert C != 0, "Denominator cannot be zero"
        assert max_time > 0, "Maximum time should be positive."
        
        self.alpha = alpha
        self.beta = beta
        self.A = A
        self.B = B
        self.C = C
        self.omega = omega
        self.method = method
        self.x_points = x_points
        self.initial_value = initial_value
        self.max_time = max_time
        self.CFL = CFL

        # set up for solution
        self.N_grid = np.size(x_points)
        self._solution_t = []
        self._solution_y = []

        # set up surrounding function using the class in Q1
        self.Surrounding_function_S = Surrounding_function_S(self.A, self.B, self.omega, self.C)
    
        # You need to write this function above for this to work!
        if method == 'finite_difference':
            test_x_points = np.linspace(x_points[0] , x_points[-1],  np.size(x_points))
            assert np.all(test_x_points == x_points), "Wrong input! Need equally spacing points for using finite difference method"
            self.D2_matrix = get_FD_second_derivative_matrix(self.N_grid - 1)
        elif method == "pseudospectral":
            # Test whether the input is Gauss Lebato nodes
            test_u_points = np.cos(np.pi * np.arange(np.size(x_points)) / (np.size(x_points) - 1))
            test_x_points = (x_points[0] - x_points[-1])/2.0 * test_u_points + (x_points[0] + x_points[-1])/2.0
            assert np.all(test_x_points == x_points), "Wrong input! Need Gauss Lebato nodes for using pseudospectral method"
            
            self.D2_matrix = get_pseudospectral_second_derivative_matrix(self.N_grid - 1)
        else:
            assert False, "Requested derivative method not implemented"

        
        # Rescale for the derivative matrix for the interval we are looking at
        b_minus_a = self.x_points[-1] - self.x_points[0]
        self.D2_matrix *= 4.0 / b_minus_a / b_minus_a

    def calculate_dydt(self, t, current_state) :
        
        # Just for readability
        dTdt = np.zeros_like(current_state)
        
        # Now actually work out the time derivatives
        dTdt[:] =  self.alpha * np.dot(self.D2_matrix, current_state) - \
                   self.beta * ( (current_state ** 4) - ((self.Surrounding_function_S.spatial_profile(self.x_points, t)) ** 4) )
           
        # Zero the derivatives at the end for stability 
        # (especially important in the pseudospectral method)
        dTdt[0] = 0.0
        dTdt[1] = 0.0
        dTdt[self.N_grid-1] = 0.0
        dTdt[self.N_grid-2] = 0.0
        
        return dTdt       
 
    def integrate_using_solve_ivp(self) :
        
        t_solution = np.linspace(0.0, self.max_time, 100)
        solution = solve_ivp(self.calculate_dydt, [0, self.max_time],
                        self.initial_value, t_eval=t_solution)
        
        self._solution_t = solution.t
        self._solution_y = solution.y
        
        self._solution_solved = True
        
        return solution.t, solution.y
    
    
    def integrate_using_midpoint(self) :
        
        assert self.method == "finite_difference", 'need to use FDs'
        dx = self.x_points[1] - self.x_points[0] # assume constant using FDs
        # Using half the CFL condition in practise just to be sure it is stable
        delta_t = self.CFL * dx
        num_timesteps = int(self.max_time / delta_t) + 1
        t_solution = np.linspace(0.0, self.max_time, num_timesteps)
        y_solution = np.zeros([self.N_grid, num_timesteps])
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:, itime] = self.initial_value
                old_time = time
            else :
                dydt_at_t = self.calculate_dydt(old_time, y_solution[:, itime-1])
                
                # Take half step
                y_t_plus_half = y_solution[:, itime-1] + 0.5 * dydt_at_t * delta_t
                t_plus_half = old_time + 0.5 * delta_t
                dydt_at_t_plus_half = self.calculate_dydt(t_plus_half, y_t_plus_half)
                
                # Take full step
                y_solution[:, itime] = (y_solution[:, itime-1] + dydt_at_t_plus_half * delta_t) 
                old_time = time 

        self._solution_t = t_solution
        self._solution_y = y_solution
        
        return t_solution, y_solution
```

We start to solve the PDE with the RK2 time integrate method. The initial is obtained using the `t_function_at_time_0` defined in Q2c. For an initial value problem, achieving a stable solution usually requires satisfying the “CFL condition” on the time step of the form: $\Delta t = \lambda \Delta x^{2}$ for parabolic equations. By physical arguments, we expect $\lambda \le 1/\alpha$ for heat equation. However, in the result, we can observe that the CFL value should be approximately $1/20\alpha$ for stability by experiments.

If the problem contains several frequencies with very different magnitudes, two main challenges occur:

2. **Accuracy** : The RK2 method may hard to capture high-frequency oscillations, especially when $\Delta x$ or $\Delta t$ is not small enough.

```python
class EllipticSolver(HeatEquationPDE):
    """
    Extends the HeatEquationPDE class with additional attributes and modified methods.
    """
    def __init__(self, x_points, initial_value, max_time, CFL=1.0, method='finite_difference',
                 alpha=0.5, beta=0.0001, A=20.0, B=1.0, C=1.5, omega=0.5):
        # Initialize the base class attributes
        super().__init__(x_points, initial_value, max_time, CFL, method, alpha, beta, A, B, C, omega)
    
    def integrate_using_solve_ivp(self, a_num_steps=100):
        assert a_num_steps > 0, "Number of steps for simulation needs to be positive"
        assert type(a_num_steps) == int, "Number of steps needs to be integer"
        
        t_solution = np.linspace(0.0, self.max_time, a_num_steps)
        print(t_solution.size)
        delta_t = self.max_time / (a_num_steps - 1)
        
        solution = solve_ivp(self.calculate_dydt, [0, self.max_time],
            self.initial_value, t_eval=t_solution)
        
        self._solution_t = solution.t
        self._solution_y = solution.y
        
        self.check_solution_converged(a_num_steps, delta_t)
        
        return solution.t, solution.y
    

    def check_solution_converged(self, a_num_steps, a_delta_t) :
     
        # Check whether the solution is converged
        last_but_one_solution = self._solution_y[:,a_num_steps-2]
        last_solution = self._solution_y[:,a_num_steps-1]
        
        change_at_last_step = last_solution - last_but_one_solution
        final_difference = np.linalg.norm(change_at_last_step, 2)/a_delta_t
        is_converged = (final_difference < 0.05)
        print("The error in the final solution is", final_difference) 
        assert is_converged, 'solution not converged, try increasing max_time or alpha'
        
        self._solution_solved = True


    def integrate_using_implicit_method(self, a_num_steps=1000) :
        assert a_num_steps > 0, "Number of steps for simulation needs to be positive"
        assert type(a_num_steps) == int, "Number of steps needs to be integer"
        
        t_solution = np.linspace(0.0, self.max_time, a_num_steps)
        y_solution = np.zeros([self.N_grid, a_num_steps])
        delta_t = self.max_time / (a_num_steps - 1)

        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:,itime] = self.initial_value
            else :
                y_of_t_old   = y_solution[:,itime-1]

                # Remember we have to copy if we don't want to actually amend the elements of y_of_t_old
                y_of_t_guess = np.copy(y_of_t_old)

                # Implement the iterative scheme described in the lecture
                error = 10.0
                error_threshold = 1e-1
                while (error > error_threshold) :

                    # Get the value of dydt at the guessed value
                    dydt_at_t_next = self.calculate_dydt(time, y_of_t_guess)

                    # Get the new guess using the Euler update rule
                    y_of_t_new = y_of_t_old + delta_t * dydt_at_t_next

                    # If the new guess is the same as the old one we have converged
                    error = np.linalg.norm(y_of_t_guess - y_of_t_new)
                    y_of_t_guess = y_of_t_new

                    # Check it is converged
                    assert error < 20.0, 'Non linear iterations not converging!'

                # Now (assuming it has converged) assign the 
                # value we found to the solution and continue the time iteration
                # Enforce the boundary conditions
                y_of_t_guess[0] = 20.0
                y_of_t_guess[self.N_grid-1] = 20.0
                y_solution[:,itime] = y_of_t_guess

        self._solution_t = t_solution
        self._solution_y = y_solution
        self.check_solution_converged(a_num_steps, delta_t)

        return t_solution, y_solution

```
